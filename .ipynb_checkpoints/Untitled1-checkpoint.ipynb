```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
```

```python
iris_dataset = load_iris()
```

```python
print(f"Ключи iris_dataset: \n{iris_dataset.keys()}")
```

```python
print(f"Первые пять строк массива data: \n {iris_dataset['data'][:5]}")
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    iris_dataset['data'], 
    iris_dataset['target'],
    random_state = 0)
```

```python
knn = KNeighborsClassifier(n_neighbors = 1)
```

```python
knn.fit(X_train, y_train)
```

```python
import numpy as np
```

```python
X_new = np.array([[5, 2.9, 1, 0.2]])
print(f"форма массива X_new: {X_new.shape}")
```

```python
prediction = knn.predict(X_new)
print("Прогноз: {}".format(prediction))
print("Спрогнозированная метка: {}".format(iris_dataset['target_names'][prediction]))
```

```python
y_pred = knn.predict(X_test)
print("Прогнозы для тестового набора: \n {}".format(y_pred))
print("Правильность на тестовом наборе: {:.2f}".format(np.mean(y_pred == y_test)))
```

```python
print("Правильность на тестовом наборе: {:.2f}".format(knn.score(X_test, y_test)))
```

```python
iris_dataframe = pd.DataFrame(X_train, columns = iris_dataset.feature_names)
grr = pd.plotting.scatter_matrix(iris_dataframe, c = y_train, alpha= .8)
grr
```
