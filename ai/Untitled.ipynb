```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("train.csv")
```

```python
df.head()
```

```python
average_age_by_class = df.groupby('Pclass')['Age'].mean()
average_age_by_class
```

```python
plt.hist(df['Age'].dropna(), bins = 20, edgecolor = 'red')
plt.xlabel('Возраст')
plt.ylabel('Частота')
plt.title('Распределение возраста пассажиров')

plt.show()
```

```python
gender_counts = df['Sex'].value_counts()

plt.pie(gender_counts, labels = gender_counts.index, autopct='%1.2f%%', startangle = 330)
plt.axis('equal')
plt.title('Соотношение мужчин и женщин на борту')
plt.show()
```

```python
sns.countplot(x = 'Embarked', hue='Survived', data = df)
plt.xlabel('Порт посадки')
plt.ylabel('Число пассажиров')
plt.title('Выживаемость в зависимости от порта посадки')
plt.legend(title = 'Survived', labels = ['не выжил', 'Выжил'])
plt.show()
```

```python
sns.kdeplot(df[df['Survived'] == 1]['Age'], label = 'Выжил', fill = True)

sns.kdeplot(df[df['Survived'] == 0]['Age'], label = 'Не выжил', fill = True)

plt.xlabel('Возраст')
plt.ylabel('Плотность')

plt.title('Плотность распределения возраста с разделением на выживших и невыживших')

plt.legend()
plt.show()
```

```python
plt.rc('figure', figsize= (15, 5))
figsize_with_subplots = (15, 15)
bin_size = 15
```

```python
fig = plt.figure(figsize = figsize_with_subplots)

fig_dims = (3, 2)

plt.subplot2grid(fig_dims, (0, 0))
df['Survived'].value_counts().plot(kind = 'bar', title = 'количество выживших и погибших')

plt.subplot2grid(fig_dims, (0, 1))
df['Pclass'].value_counts().plot(kind='bar', title = 'класс пассажиров')

plt.subplot2grid(fig_dims, (1, 0))
df['Sex'].value_counts().plot(kind='bar', title='пол пассажиров')
plt.xticks(rotation = 0)

plt.subplot2grid(fig_dims, (1, 1))
df['Embarked'].value_counts().plot(kind = 'bar', title = 'порт посадки')

plt.subplot2grid(fig_dims, (2, 0))
df['Age'].hist()
plt.title('возраст пассажиров')

plt.show()
```
