```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
iris_dataset = load_iris()
```

```python
print(f"Ключи iris_dataset: \n{iris_dataset.keys()}")
```

```python
print(f"Первые пять строк массива data: \n {iris_dataset['data'][:5]}")
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    iris_dataset['data'], 
    iris_dataset['target'],
    random_state = 0)
```

```python
knn = KNeighborsClassifier(n_neighbors = 1)
```

```python
knn.fit(X_train, y_train)
```

```python
print(f"форма массива X_train: \n {X_train.shape}")
print(f"форма массива y_train: \n {y_train.shape}")
```

```python
print(f"форма массива X_test: \n {X_test.shape}")
print(f"форма массива y_test: \n {y_test.shape}")
```

```python
import numpy as np
```

```python
X_new = np.array([[5, 2.9, 1, 0.2]])
print(f"форма массива X_new: {X_new.shape}")
```

```python
prediction = knn.predict(X_new)
print("Прогноз: {}".format(prediction))
print("Спрогнозированная метка: {}".format(iris_dataset['target_names'][prediction]))
```

```python
y_pred = knn.predict(X_test)
print("Прогнозы для тестового набора: \n {}".format(y_pred))
print("Правильность на тестовом наборе: {:.2f}".format(np.mean(y_pred == y_test)))
```

```python
print("Правильность на тестовом наборе: {:.2f}".format(knn.score(X_test, y_test)))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

# Загружаем данные (замените эту часть кода своими данными)
data = load_iris()
X = data.data

# Применяем PCA
pca = PCA(n_components=2)  # Указываем, что хотим получить 2 компоненты
X_pca = pca.fit_transform(X)

# Рисуем диаграмму рассеяния
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=data.target)
plt.xlabel('Главная компонента 1')
plt.ylabel('Главная компонента 2')
plt.show()
```
