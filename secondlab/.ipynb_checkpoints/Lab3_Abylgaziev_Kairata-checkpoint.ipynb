# 1.	Визуализация данных с помощью matplotlib. Создание графиков с использованием «научной графики».

```python
from pylab import *
import matplotlib.pyplot as plt
%matplotlib inline
```

### 2.	Построить график произвольной квадратичной функции, масштаб задать произвольно. Сделать настройки подписей осей, заголовка фигуры 

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-5, 5, 100)
y =  x**2 + 2 * x + 1
plt.figure(figsize=(8, 6 )) 
plt.plot(x, y, label=f'x^2 + 2x + 1', color='r') 
plt.xlabel('Ось X', color = 'r')  
plt.ylabel('Ось Y', color = 'r')  
plt.title('График квадратичной функции') 
plt.grid(True)
plt.legend()
plt.show()
```

### 3.	Построить график произвольной гармонической функции, масштаб задать произвольно. Сделать настройки подписей осей, заголовка фигуры

```python
x = np.linspace(-6, 2 * np.pi, 100)
y =  np.cos(x)
plt.figure(figsize=(8, 6)) 
plt.plot(x, y, label=f'cos( x )', color='b')
plt.xlabel('Ось X',  color='b') 
plt.ylabel('Ось Y',  color='b') 
plt.title('График гармонической функции')  
plt.legend()
plt.grid(True) 
plt.show()
```

### 4.	Вывести графики п.2 и 3 в разные поля при помощи функции subplot()

```python

x = np.linspace(-5, 5, 100)
y = x**2 + 2 * x + 1

x_harmonic = np.linspace(-6, 2 * np.pi, 100)
y_harmonic = np.cos(x)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)  
plt.plot(x, y, label=f'x^2 + 2x + 1', color='r')
plt.xlabel('Ось X', color = 'r')  
plt.ylabel('Ось Y', color = 'r')  
plt.title('График квадратичной функции')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)  
plt.plot(x_harmonic, y_harmonic, label=f'cos( x )', color='b')
plt.xlabel('Ось X',  color='b') 
plt.ylabel('Ось Y',  color='b') 
plt.title('График гармонической функции')  
plt.legend()
plt.grid(True)

plt.tight_layout()  
plt.show()
```

### 5.	Построить графики п.2 и 3 в одной координатной плоскости, сделать подписи графиков, указать легенду.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 100)
y = x**2 + 2 * x + 1

x_harmonic = np.linspace(-6, 2 * np.pi, 100)
y_harmonic = np.cos(x)

plt.figure(figsize=(10, 6))

plt.plot(x, y, label=f'x^2 + 2x + 1', color='r')
plt.plot(x_harmonic, y_harmonic, label=f'cos( x )', color='b')


plt.xlabel('Ось X') 
plt.ylabel('Ось Y') 
plt.title('Графики квадратичной и гармонической функций')

plt.legend()
plt.grid(True)
plt.show()
```

```python
t = np.linspace(0, 4*np.pi, 100)
x_parametric = np.cos(3*t)
y_parametric = np.sin(2*t)
plt.figure(figsize=(8, 6))
plt.plot(x_parametric, y_parametric)
plt.xlabel('Ось X') 
plt.ylabel('Ось Y') 
plt.title('График параметрической функции')
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
years = [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
min_income = [4341.15, 4599.21, 4981.51, 5182.99, 4794.34, 4900.79, 4792.54, 4806.32, 5358.53, 6268.31, 7178.32]
plt.figure(figsize=(10, 6))
plt.bar(years, min_income)
plt.xlabel('Год')
plt.ylabel('Прожиточный минимум')
plt.title('Столбиковая диаграмма прожиточного минимума в КР (2012-2022)')

plt.tight_layout()
plt.show()
```

```python
categories = ['2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022']
min_income = [4341.15, 4599.21, 4981.51, 5182.99, 4794.34, 4900.79, 4792.54, 4806.32, 5358.53, 6268.31, 7178.32]
plt.figure(figsize=(8, 8))
plt.pie(min_income, labels=categories, autopct='%1.1f%%')
plt.title('Круговая диаграмма прожиточного минимума в КР (2012-2022)')

plt.tight_layout()
plt.show()
```

### 9.	Построить график произвольной каркасной поверхности.

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(x) + np.cos(y)
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')
ax.set_xlabel('Ось X')
ax.set_ylabel('Ось Y')
ax.set_zlabel('Ось Z')

plt.title('Каркасная поверхность')
plt.show()
```
