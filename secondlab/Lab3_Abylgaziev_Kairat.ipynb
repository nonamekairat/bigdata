```python
import matplotlib.pyplot as plt
import numpy as np 

# Turn on graphing in the notebook
%matplotlib inline
```

### 2 Построить график квадратичной функции

```python
x = np.linspace(-5, 5, 100)
y = x**2
plt.plot(x, y)
plt.xlabel('X-ось')
plt.ylabel('Y-ось')
plt.title('График квадратичной функции')

plt.tight_layout()
plt.show()
```

### 3 График гармонической функции

```python
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
plt.plot(x, y)
plt.xlabel('X-ось')
plt.ylabel('Y-ось')
plt.title('График гармонической функции')

plt.tight_layout()
plt.show()
```

### 4 вывести графики п.2 и 3 в разные поля при помощи функции subplot()

```python
plt.subplot(1, 2, 1)
x = np.linspace(-5, 5, 100)
y_harmonic = np.sin(x)
plt.plot(x, y_harmonic)
plt.xlabel('X-ось')
plt.ylabel('Y-ось')
plt.title('График гармонической функции')

plt.subplot(1, 2, 2)
x = np.linspace(-5, 5, 100)
y_quadratic = x**2
plt.plot(x, y_quadratic)
plt.xlabel('X-ось')
plt.ylabel('Y-ось')
plt.title('График квадратичной функции')

plt.tight_layout()
plt.show()
```

### 5 Построить графики п.2 и 3 в одной координатной плоскости, сделать подписи графиков, указать легенду.

```python
x = np.linspace(-5, 5, 100)
y_harmonic = np.sin(x)
y_quadratic = x**2
plt.figure(figsize=(8, 4))
plt.plot(x, y_harmonic, label='Гармоническая функция')
plt.plot(x, y_quadratic, label='Квадратичная функция')
plt.xlabel('X-ось')
plt.ylabel('Y-ось')
plt.title('Графики функций')
plt.legend()

plt.tight_layout()
plt.show()
```

### 6 Построить график по функции, заданный параметрически

```python
t = np.linspace(0, 4*np.pi, 100)
x_parametric = np.cos(3*t)
y_parametric = np.sin(2*t)
plt.figure(figsize=(8, 4))
plt.plot(x_parametric, y_parametric)
plt.xlabel('X-ось')
plt.ylabel('Y-ось')
plt.title('График параметрической функции')

plt.tight_layout()
plt.show()
```

### 7 построить столбиковую диаграмму

```python
years = [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
min_income = [4341.15, 4599.21, 4981.51, 5182.99, 4794.34, 4900.79, 4792.54, 4806.32, 5358.53, 6268.31, 7178.32]
plt.bar(years, min_income)
plt.xlabel('Год')
plt.ylabel('Прожиточный минимум')
plt.title('Столбиковая диаграмма прожиточного минимума в КР (2012-2022)')

plt.tight_layout()
plt.show()
```

### 8 построить круговой график

```python
categories = ['2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022']
min_income = [4341.15, 4599.21, 4981.51, 5182.99, 4794.34, 4900.79, 4792.54, 4806.32, 5358.53, 6268.31, 7178.32]
plt.pie(min_income, labels=categories, autopct='%1.1f%%')
plt.title('Круговая диаграмма прожиточного минимума в КР (2012-2022)')

plt.tight_layout()
plt.show()
```

### 9 построить график произвольной каркасной поверхности

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(x) + np.cos(y)
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')
ax.set_xlabel('X-ось')
ax.set_ylabel('Y-ось')
ax.set_zlabel('Z-ось')
plt.title('Каркасная поверхность')
```
